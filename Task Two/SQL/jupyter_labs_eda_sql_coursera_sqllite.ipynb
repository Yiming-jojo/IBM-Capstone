## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
# Write the SQL query
query = "SELECT DISTINCT Launch_Site FROM SPACEXTBL;"

# Execute the query and fetch the results
unique_launch_sites = pd.read_sql_query(query, con)

unique_launch_sites
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA'


```python
# SQL query to select 5 records where the launch sites start with 'CCA'
query = "SELECT * FROM SPACEXTBL WHERE Launch_Site LIKE 'CCA%' LIMIT 5;"

# Execute the query and fetch the records
records = pd.read_sql_query(query, con)

records
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# SQL query to calculate the total payload mass for NASA (CRS)
query = "SELECT SUM(Payload_Mass__kg_) as Total_Payload_Mass FROM SPACEXTBL WHERE Customer LIKE 'NASA (CRS)';"

# Execute the query and store the result in a DataFrame
total_payload_mass = pd.read_sql_query(query, con)
```

```python
total_payload_mass
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
# SQL query to calculate the average payload mass for the booster version F9 v1.1
query = "SELECT AVG(Payload_Mass__kg_) as Average_Payload_Mass FROM SPACEXTBL WHERE Booster_Version = 'F9 v1.1';"

# Execute the query and fetch the result
average_payload_mass = pd.read_sql_query(query, con)
```

```python
average_payload_mass
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_


```python
# SQL query to find the first successful ground pad landing
query = """
SELECT Date FROM SPACEXTBL
WHERE Landing_Outcome = 'Success (ground pad)'
ORDER BY Date
LIMIT 1;
"""

# Execute the query and fetch the result
first_successful_landing_date = pd.read_sql_query(query, con)
```

```python
first_successful_landing_date
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
# SQL query to find boosters with successful drone ship landings and payload mass between 4000 and 6000 kg
query = """
SELECT Booster_Version FROM SPACEXTBL
WHERE Landing_Outcome = 'Success (drone ship)'
AND Payload_Mass__kg_ > 4000 AND Payload_Mass__kg_ < 6000;
"""

# Execute the query and fetch the result
boosters = pd.read_sql_query(query, con)
```

```python
boosters
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
# SQL query to count the total number of successful and failed mission outcomes
query = """
SELECT Mission_Outcome, COUNT(*) as Total
FROM SPACEXTBL
GROUP BY Mission_Outcome;
"""

# Execute the query and fetch the result
mission_outcomes = pd.read_sql_query(query, con)
```

```python
mission_outcomes
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
# SQL query to find the booster_versions that have carried the maximum payload mass
query = """
SELECT Booster_Version FROM SPACEXTBL
WHERE Payload_Mass__kg_ = (SELECT MAX(Payload_Mass__kg_) FROM SPACEXTBL);
"""

# Execute the query and fetch the result
booster_versions = pd.read_sql_query(query, con)
```

```python
booster_versions
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
# SQL query to find the specific records for 2015 with failure landing outcomes at the drone ship
query = """
SELECT
    SUBSTR(Date, 6, 2) AS Month,
    Landing_Outcome,
    Booster_Version,
    Launch_Site
FROM
    SPACEXTBL
WHERE
    SUBSTR(Date, 1, 4) = '2015'
    AND Landing_Outcome LIKE 'Failure (drone ship)'
ORDER BY
    Date;
"""

# Execute the query and fetch the result
records_2015 = pd.read_sql_query(query, con)
```

```python
records_2015
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
# SQL query to rank the count of landing outcomes between the specified dates
query = """
SELECT Landing_Outcome, COUNT(Landing_Outcome) as Outcome_Count
FROM SPACEXTBL
WHERE Date >= '2010-06-04' AND Date <= '2017-03-20'
GROUP BY Landing_Outcome
ORDER BY Outcome_Count DESC;
"""

# Execute the query and fetch the result
landing_outcomes_ranked = pd.read_sql_query(query, con)

# Close the database connection
con.close()

landing_outcomes_ranked
```
